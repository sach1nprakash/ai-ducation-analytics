```python
from deepface import DeepFace
import os
import shutil

# Path to the folder containing images
input_folder_path = "./Dataset/dataset-split/Test/Bored"
output_folder_path_young = "./Dataset/dataset-bias/Test/Bored/Age/Young"
output_folder_path_mid = "./Dataset/dataset-bias/Test/Bored/Age/Mid"
output_folder_path_senior = "./Dataset/dataset-bias/Test/Bored/Age/Senior"

# Create output folders if they don't exist
os.makedirs(output_folder_path_young, exist_ok=True)
os.makedirs(output_folder_path_mid, exist_ok=True)
os.makedirs(output_folder_path_senior, exist_ok=True)

# Iterate over each file in the folder
for file in os.listdir(input_folder_path):
    file_path = os.path.join(input_folder_path, file)

    # Analyze the image
    # objs = DeepFace.analyze(img_path=file_path, actions=['gender'], enforce_detection=False)
    objs = DeepFace.analyze(img_path=file_path, actions=['age'], enforce_detection=False)

    # Determine the gender
    # dominant_gender = objs[0]['dominant_gender']
    dominant_age = objs[0]['age']


    # Copy the file to the appropriate output folder
    # if dominant_gender == "Man":
    #     shutil.copy(file_path, output_folder_path_man)
    # else:
    #     shutil.copy(file_path, output_folder_path_woman)

    # Copy the file to the appropriate output folder
    if dominant_age in range(27):
        shutil.copy(file_path, output_folder_path_young)
    elif dominant_age in range(28, 55):
        shutil.copy(file_path, output_folder_path_mid)
    else:
        shutil.copy(file_path, output_folder_path_senior)

# Print results
print("Age distribution:")
print(f"Young: {len(os.listdir(output_folder_path_young))}")
print(f"Mid: {len(os.listdir(output_folder_path_mid))}")
print(f"Senior: {len(os.listdir(output_folder_path_senior))}")
```
